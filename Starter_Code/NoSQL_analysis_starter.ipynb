### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$gte':20}}
hresults = establishments.find(query)

# Use count_documents to display the number of documents in the result
num_results = establishments.count_documents(query)
print("Number of docs with a hygiene score >20: ", num_results)
# Display the first document in the results using pprint
if num_results > 0:
    first_result = hresults[0]
    pprint(first_result)
else:
    print("No documents with hygiene score > 20 found.")
```

```python
import pandas as pd
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hresults)
# Display the number of rows in the DataFrame
num_rows = hygiene_df.shape[0]
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue':{'$gte': 4},
         'AddressLine3':'London'}

londonresults = establishments.find(query)


# Use count_documents to display the number of documents in the result
nums_results = establishments.count_documents(query)
print('Number of stores with a rating >= 4 and it is in London: ', nums_results)
# Display the first document in the results using pprint
if nums_results > 0:
    first_result = londonresults[0]
    pprint(first_result)
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(londonresults)
# Display the number of rows in the DataFrame
num_rows = london_df.shape[0]
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score
import pymongo
import pandas as pd
degree_search = 0.01
latitude = 51.4934
longitude = 0.0098

query = {'RatingValue':5, 
         'AddressLine3':'Greenwich',
         'geocode.latitude': {
             '$gte': latitude - degree_search,
             '$lte': latitude + degree_search
         },
         'geocode.longitude': {
             '$gte': longitude - degree_search,
             '$lte': longitude + degree_search
         }
}
sort =  [('scores.Hygiene', pymongo.ASCENDING)]

# Print the results
results = establishments.find(query).sort(sort).limit(5)

result_list = list(results) 


```

```python
# Convert result to Pandas DataFrame
panang_df = pd.DataFrame(result_list)
panang_df.head(3)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
pipeline = [
    {
        '$match': {
            'scores.Hygiene':0
        }
    },
    {
        '$group': {
            '_id': '$LocalAuthorityName',
            'count': {'$sum': 1}
        }
    },
    {
        '$sort' : {
            'count':-1
        }
    }
]
# Print the number of documents in the result
result = establishments.aggregate(pipeline)
result_count = 0
for doc in result:
    result_count += 1

print(f"Number of documents in the result: {result_count}")
result = establishments.aggregate(pipeline)
# Print the first 10 results


agg_result = list(result)    
```

```python
# Convert the result to a Pandas DataFrame
agg_df = pd.DataFrame(agg_result)
# Display the number of rows in the DataFrame
row_nums = agg_df.shape[0]
print(f"Number of rows in the Dataframe: {row_nums}")
# Display the first 10 rows of the DataFrame
agg_df.head(10)
```
